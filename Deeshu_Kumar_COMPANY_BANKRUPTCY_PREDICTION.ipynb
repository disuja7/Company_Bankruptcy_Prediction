```python
# Importing libraries
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
# Creating an SFS object
sfs = SFS(estimator=lr,       # Using logistic regression as our classifier
          k_features=(1,20),  # Considering any feature combination between 1 and 20
          forward=True,       # Set forward to True when we want to perform SFS
          scoring='precision', # The metric to use to evaluate the classifier is precision 
          cv=3)               # The number of cross-validations to perform is 3
# Train SFS with our dataset
sfs = sfs.fit(df.drop(['Bankrupt?'],axis=1), df['Bankrupt?'])
# Print the results
print('Best precision: %.2f' % sfs.k_score_)   
print('Best subset (indices):', sfs.k_feature_idx_)                                                     
print('Best subset (corresponding names):', sfs.k_feature_names_)
```

```python
# Transform the dataset
X_data_new = sfs.transform(df[:])
# Print the results
print('Number of features before transformation: {}'.format(df.shape[1]))
print('Number of features after transformation: {}'.format(X_data_new.shape[1]))
```

## **Model Selection**

Here our task is to predict whether the company is Bankrupt or not. This task can be done using a classification model.
Now the question is which classification model should we use?

And the answer for this question depends on the business requirement. Since we know that as we increase the accuracy of prediction using complex models, the explainability of the model reduces, which might create an issue for the stakeholders.

Considering the above situations we can have three choices-


*   Simple model with low accuracy but high explainability
*   Complex model with high accuracy but low explainability
*   A compromised model between the above two cases
